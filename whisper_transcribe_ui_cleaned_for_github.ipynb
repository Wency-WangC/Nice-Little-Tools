```python
import whisper
from docx import Document

audio_file = r"PATH_TO_AUDIO.m4a"
word_file_path = r"OUTPUT.docx"

# 加载 Whisper 模型
model = whisper.load_model("turbo")

# 转录音频或视频文件
result = model.transcribe(audio_file, verbose=False)

# 获取转录的文本
transcription_text = result["text"]

doc = Document()
doc.add_paragraph(transcription_text)

# 添加标题
doc.add_heading('Transcription Result', 0)

# 添加转录文本
doc.add_paragraph(transcription_text)

# 保存文档
doc.save(word_file_path)

print(f"转录文本已保存为：{word_file_path}")
```

```python
try:
    import whisper
except ModuleNotFoundError:
    raise ModuleNotFoundError("Whisper module is not installed. Please install it using 'pip install openai-whisper'.")

try:
    import easygui
except ModuleNotFoundError:
    raise ModuleNotFoundError("EasyGUI module is not installed. Please install it using 'pip install easygui'.")

import os
import tkinter as tk
from tkinter import simpledialog, messagebox
from docx import Document
from tqdm import tqdm

def transcribe_audio(file_path, output_path):
    """使用Whisper转录音频，并保存到Word文档"""
    model = whisper.load_model("turbo")
    print("当前处理文件路径:", file_path)
    result = model.transcribe(file_path, verbose=False)
    transcription_text = result["text"]
    
    doc = Document()
    doc.add_heading('Transcription Result', 0)
    doc.add_paragraph(transcription_text)
    doc.save(output_path)

def main():
    root = tk.Tk()
    root.withdraw()  # 隐藏主窗口
    
    # 选择音频文件所在的目录
    folder_path = easygui.diropenbox(title="请选择包含音频文件的文件夹")
    if not folder_path:
        messagebox.showinfo("提示", "未选择文件夹，程序退出。")
        return
    
    # 获取音频文件
    audio_extensions = (".mp4", ".mp3", ".wav", ".m4a")
    audio_files = [f for f in os.listdir(folder_path) if f.endswith(audio_extensions)]
    
    if not audio_files:
        messagebox.showinfo("提示", "未找到音频文件，程序退出。")
        return
    
    # 让用户选择要转录的音频
    selected_files = easygui.multchoicebox("请选择要转录的音频文件", "音频选择", audio_files)
    if not selected_files:
        messagebox.showinfo("提示", "未选择音频文件，程序退出。")
        return
    
    # 处理选中的音频
    for file in tqdm(selected_files, desc="转录进度", unit="file", leave=True, ncols=100):
        file_path = os.path.join(folder_path, file)
        default_output_path = os.path.join(folder_path, f"{os.path.splitext(file)[0]}.docx")
        
        # 询问是否修改文件名
        new_name = simpledialog.askstring("文件名", f"输入{file}的保存文件名(不输入则使用默认名):")
        output_path = default_output_path if not new_name else os.path.join(folder_path, f"{new_name}.docx")
        
        transcribe_audio(file_path, output_path)
    
    messagebox.showinfo("完成", "所有选定文件已成功转录！")

if __name__ == "__main__":
    main()
```
